```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetics = load_diabetes()
```

```python
column_name=diabetics.feature_names
df=pd.DataFrame(diabetics.data)
df.columns=column_name
df.head()
```

```python
sns.boxplot(df.bmi)
```

```python
df.shape
```

```python

q1=df['bmi'].quantile(0.25)
```

```python
q1
```

```python
q3=df['bmi'].quantile(0.75)
```

```python
q3
```

```python
IQR=q3-q1
```

```python
IQR
```

```python
l=q1-1.5*IQR
u=q3+1.5*IQR
ub=np.where(df['bmi']>=u)[0]
lb=np.where(df['bmi']<=l)[0]
```

```python
L
```

```python
u
```

```python
df['bmi']=df['bmi'].drop(index = ub)
```

```python
df['bmi']=df['bmi'].drop(index = lb)
```
