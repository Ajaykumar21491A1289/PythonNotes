```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.5])
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('oop plot')
axes.plot(x,y)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,1,1])
axes2=fig.add_axes([0.3,0.6,0.4,0.3])
axes1.set_title('outer plot')
axes2.set_title('inner plot')
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
#sub plots
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.show()
```
